###**PROCESS THE IMBALANCED DATA**

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

data_matrix = df.values

# X: feature matrix (for this analysis, we exclude the Time variable from the dataset)
X = data_matrix[:, 1:30]

# y: labels vector
y = data_matrix[:, 30]


# print the shape of the features matrix and the labels vector
print('X.shape=', X.shape, 'y.shape=', y.shape)
# Create an imbalanced data processing pipeline

resampling_pipeline = Pipeline([
    ('oversample', SMOTE(sampling_strategy=0.5)),  # Increase the minority class to 50% of the majority class
    ('undersample', RandomUnderSampler(sampling_strategy=1.0))  # Decrease the majority class to the same size as the minority class
])

X_resampled, y_resampled = resampling_pipeline.fit_resample(X, y)
```

###**TRAIN THE MODEL**

```python
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
# Assuming X_train and y_train are your training features and labels
# Create a Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust the number of estimators (trees) as needed

# Train the Random Forest Classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on test data
predictions = rf_classifier.predict(X_test)

```

```python
print("Length of y:", len(y))
print("Length of predictions:", len(predictions))
```

###**ACCURACY SCORES USING DIFFERENT METRICS**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
# Accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Precision
precision = precision_score(y_test, predictions)
print("Precision:", precision)

# Recall
recall = recall_score(y_test, predictions)
print("Recall:", recall)

# F1-score
f1 = f1_score(y_test, predictions)
print("F1-score:", f1)

# ROC AUC Score (only applicable for binary classification with probability scores)
roc_auc = roc_auc_score(y_test, predictions)
print("ROC AUC Score:", roc_auc)
```
